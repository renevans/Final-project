```python
#Code for the appartments

%%time
# Using selenium:

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
  
# for holding the resultant list
element_list = []
# address=[]
# postal_code = []
# listed_price=[]
# living_area=[]
# plot_size=[]
# number_of_rooms=[]
# agent=[] 
      
    
for page in range(657,658): # started at page 1 and compiled in batches
    
    page_url = "https://www.funda.nl/en/koop/amsterdam/verkocht/appartement/sorteer-afmelddatum-af/p" + str(page) + "/"
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get(page_url)
    time.sleep(3)
#     wait_time = randint(1,4)
    driver.implicitly_wait(4)
    address = driver.find_elements_by_css_selector("a:nth-child(1) > h2")
    postal_code = driver.find_elements_by_css_selector("a:nth-child(2) > h4")
    listed_price= driver.find_elements_by_css_selector("div.search-result-info.search-result-info-price > span") # listed price
    living_area= driver.find_elements_by_css_selector("li:nth-child(1) > span:nth-child(1)") # living area
#   plot_size.append(soup.select("li:nth-child(1) > span:nth-child(2)")[i].get_text().strip()) # plot size
    number_of_rooms= driver.find_elements_by_css_selector("div:nth-child(3) > ul > li:nth-child(2)") # number of rooms
    agent= driver.find_elements_by_css_selector("span.search-result-makelaar-name") # agent   
    
#     time.sleep(3)
    
#     description = driver.find_elements_by_class_name("description")
#     rating = driver.find_elements_by_class_name("ratings")

    for i in range(len(address)):
        element_list.append([address[i].text, postal_code[i].text, listed_price[i].text, living_area[i].text,  number_of_rooms[i].text, agent[i].text])
      
#         print(element_list)
  
#closing the driver
driver.close()


```

```python
df = pd.DataFrame(element_list, columns = ['address', 'postal_code','listed_price','living_area','number_of_rooms','agent'])

```

```python
 element_list5 = []  #compiled in batches from 1 to 5
    
for url in url_list_final_sliced:
    page_url = url  
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get(page_url)
    time.sleep(3)
    html = driver.page_source
    driver.quit()
    soup = BeautifulSoup(html, 'html.parser')
    element_list5.append(soup.select("div.object-primary"))
    time.sleep(randint(8,15))  
```

```python
# element_list
```

```python
# This works to compile the df after the soup is created

address_2=[]
neighbourhood = []
listed_since=[]
date_of_sale=[]
year_of_construction=[]
number_of_rooms_detail=[]
number_of_bath_rooms=[]
bathroom_facilities=[]
number_of_stories=[]
located_at=[]
energy_label=[]
ownership_situation=[]
VVE_contribution=[]
# num_iter = len(element_list.select("h1 > span.object-header__title"))

for i in element_list5:
    try:
        address_2.append(i[0].select("h1 > span.object-header__title")[0].get_text().strip()) #address
    except:
        address_2.append("N/A")
    try:
        neighbourhood.append(i[0].select("h1 > span.object-header__subtitle.fd-color-dark-3")[0].get_text().strip()) #neighbourhood
    except:
        neighbourhood.append("N/A")
    try:   
        listed_since.append(i[0].select("dl > dd:nth-child(2)")[0].get_text().strip()) #listed_since
    except:
        listed_since.append("N/A")
    try:     
        date_of_sale.append(i[0].select("dl > dd:nth-child(4)")[0].get_text().strip()) #date_of_sale
    except:    
        date_of_sale.append("N/A")
    try:
        year_of_construction.append(i[0].select("dl:nth-child(5) > dd:nth-child(6)")[0].get_text().strip()) #year_of_construction
    except:
        year_of_construction.append("NA")
    try:
        number_of_rooms_detail.append(i[0].select("dl:nth-child(11) > dd:nth-child(2)")[0].get_text().strip()) #number_of_rooms_detail
    except: 
        number_of_rooms_detail.append("NA")
    try: 
        number_of_bath_rooms.append(i[0].select("dl:nth-child(11) > dd:nth-child(4)")[0].get_text().strip()) #number_of_bath_rooms
    except:
        number_of_bath_rooms.append("NA")
    try:
        bathroom_facilities.append(i[0].select("dl:nth-child(11) > dd:nth-child(6)")[0].get_text().strip()) #bathroom_facilities
    except:
        bathroom_facilities.append("NA")
    try:    
        number_of_stories.append(i[0].select("dl:nth-child(11) > dd:nth-child(8)")[0].get_text().strip()) #number_of_stories
    except:
        number_of_stories.append("NA")
    try:    
        located_at.append(i[0].select("dl:nth-child(11) > dd:nth-child(10)")[0].get_text().strip()) #located_at
    except:   
        located_at.append("NA")
    try:   
        energy_label.append(i[0].select("dl:nth-child(14) > dd:nth-child(2)")[0].get_text().strip()) #energy_label
    except: 
        energy_label.append("NA")
    try:  
        ownership_situation.append(i[0].select("dl:nth-child(17) > dd.object-kenmerken-group-list > dl > dd")[0].get_text().strip()) #ownership_situation
    except:
        ownership_situation.append("NA")
    try: 
        VVE_contribution.append(i[0].select("dl:nth-child(26) > dd:nth-child(6)")[0].get_text().strip()) #VVE_contribution
    except:
        VVE_contribution.append("NA")
```
