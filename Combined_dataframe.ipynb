```python
def address_new(a):
    return re.sub(r"(?<=\d)\s+", "-", a)
```

```python
def address_new2(a):
    return re.sub(r"--", "-", a)
```

```python
def address_lettersI(a):
    return re.sub(r"-I", "-1", a)
```

```python
def address_lettersII(a):
    return re.sub(r"-II", "-2", a)
```

```python
def address_lettersIII(a):
    return re.sub(r"III", "3", a)
```

```python
def address_lettersIV(a):
    return re.sub(r"IV", "4", a)
```
